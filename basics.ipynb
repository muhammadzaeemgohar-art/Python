### 📝 Challenge 1: Temperature Converter 

**Task:** Write a program that converts temperatures seamlessly between **Celsius**, **Fahrenheit**, and **Kelvin**.

---

### 📋 Requirements

1. **Temperature Input:** Prompt the user to enter a numeric temperature value.
2. **Current Unit:** Prompt the user to enter the source unit (`C`, `F`, or `K`).
3. **Target Unit:** Prompt the user to enter the desired unit (`C`, `F`, or `K`).
4. **Conversion Logic:** Correctly handle all 6 possible conversion combinations.
5. **Precision:** Print the final answer rounded to exactly **2 decimal places**.
6. **Error Handling:** Gracefully handle and report any invalid unit inputs.

---

### 💡 Quick Reference Formulas

| Conversion | Formula |
| :--- | :--- |
| **Celsius to Fahrenheit** | $F = (C \times \frac{9}{5}) + 32$ |
| **Celsius to Kelvin** | $K = C + 273.15$ |
| **Fahrenheit to Celsius** | $C = (F - 32) \times \frac{5}{9}$ |

<div style="padding: 15px; border-left: 5px solid #2196F3; background-color: #eef7fe; color: #0d47a1; border-radius: 4px;">
    <strong>ℹ️ Tip:</strong> Don't forget to test edge cases, such as entering lowercase letters (<code>c</code>, <code>f</code>, <code>k</code>) or absolute zero!
</div>

---
### Solution

````python
current_temp=float(input("Enter Temperature:"))
current_unit=(str(input("Enter the source unit C,F or K:"))).capitalize()
target_unit=(str(input("Enter target unit C,F or K:"))).capitalize()
#Celcius to Farhenhit
if ((current_unit=="C") and (target_unit =="F")):
        converted_temp=((current_temp)*(9/5))+32
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Celcius to Kelvin
elif ((current_unit=="C") and (target_unit=="K")):
    converted_temp=current_temp+273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Farhenheit to Celcius
elif ((current_unit=="F") and (target_unit=="C")):
    if current_temp==32:
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is 0")
    else:
        converted_temp=(current_temp-32)*(5/9)
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Farhenheit to Kelvin
elif ((current_unit=="F") and (target_unit=="K")):
    converted_temp=(current_temp-32)*(5/9)+273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#kelvin to Celcius
elif ((current_unit=="K") and (target_unit=="C")):
    converted_temp=current_temp-273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Kelvin to Farhenheit
elif ((current_unit=="K") and (target_unit=="F")):
    converted_temp=(current_temp-273.15)*(9/5)+32
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#For Same Units
elif (current_unit==target_unit):
    print(f"The Temperature in {current_unit} is {current_temp}")
else:
    print("Invalid Input! Try Again.")
````

## 📝 Challenge 2: BMI Calculator ⭐

**Task:** Write a program that calculates a user's Body Mass Index (BMI) based on their height and weight, and determines their weight category.

---

### 📋 Requirements

1. **User Inputs:** Prompt the user for the following three inputs:
   * **Name** (Text)
   * **Weight** in kilograms (Numeric)
   * **Height** in meters (Numeric)
2. **BMI Calculation:** Calculate the BMI using the formula below.
3. **Output Formatting:** Print a clean summary displaying:
   * The user's **Name**
   * Their calculated **BMI** rounded to exactly **2 decimal places**
   * Their official **BMI Category** based on the classification table below.

---

### 🧮 Formula & Classifications

$$BMI = \frac{\text{weight}}{\text{height}^2}$$

| BMI Range | Category |
| :--- | :--- |
| **Less than 18.5** | Underweight |
| **18.5 – 24.9** | Normal |
| **25.0 – 29.9** | Overweight |
| **30.0 or more** | Obese |

<div style="padding: 15px; border-left: 5px solid #4CAF50; background-color: #f1f8e9; color: #33691e; border-radius: 4px;">
    <strong>ℹ️ Tip:</strong> Make sure your height input handles decimal values (like <code>1.75</code>) and weight handles floats (like <code>68.5</code>) to ensure calculation accuracy!
</div>

---
### 💻Solution

````python
#Inputs
name=str(input("Enter Your Name:"))
weight=float(input("Enter Your Weight in kilograms:"))
height=float(input("Enter Your Height in meters:"))
#Calculation
bmi=weight/(height)**2
#classification
if bmi<18.5:
    category="Underweight"
elif bmi>=18.5 and bmi<=24.9:
    category="Normal"
elif bmi>=25.0 and bmi<=29.9:
    category="Overweight"
else:
    category="obese"
#Results
print(f"{'Name':<12} {name}")
print(f"{'BMI':<12} {bmi:.2f}")
print(f"{'Category':<12} {category}")
````

## 📝 Challenge 5: Electricity Bill ⭐⭐

**Task:** Write a program that calculates an electricity bill based on a slab/tier tariff structure, adds tax, and prints a final invoice.

---

### 📋 Requirements

1. **User Inputs:** Prompt the user to enter:
   * **Customer Name** (Text)
   * **Units Consumed** (Integer)
2. **Slab Rate Logic:** Calculate the base bill using a tiered system:
   * **First 100 units:** Rs. 20 per unit
   * **Next 100 units (101-200):** Rs. 30 per unit
   * **Remaining units (201+):** Rs. 40 per unit
3. **Taxation:** Apply a flat **5% tax** on the calculated base bill.
4. **Output Formatting:** Print a clean, itemized invoice showing:
   * **Customer Name**
   * **Units Consumed**
   * **Base Bill Amount** (Rs.)
   * **Calculated Tax** (Rs.)
   * **Total Bill Amount** (Base Bill + Tax) formatted to **2 decimal places**.

---

### 📥 Example Walkthrough

#### If a customer uses 250 units:
* **First 100 units:** $100 \times 20 = \text{Rs. } 2,000$
* **Next 100 units:** $100 \times 30 = \text{Rs. } 3,000$
* **Remaining 50 units:** $50 \times 40 = \text{Rs. } 2,000$
* **Base Bill:** $\text{Rs. } 7,000$
* **Tax (5%):** $\text{Rs. } 350$
* **Total Bill:** $\text{Rs. } 7,350$

<div style="padding: 15px; border-left: 5px solid #F44336; background-color: #ffebee; color: #b71c1c; border-radius: 4px;">
    <strong>⚠️ Logic Warning:</strong> Do not just multiply the total units by a single rate! If a user consumes 250 units, they do not pay Rs. 40 for all 250 units. You must calculate the cost incrementally across each bracket using conditional (<code>if-elif-else</code>) structures.
</div>

---
### 💻Solution

````python
#Input
name=str(input("Enter Your Name:"))
units_consumed=int(input("Enter Units Consumed:"))
#Calculation
if (units_consumed>0):
    if (units_consumed<=100):
        base_bill=units_consumed*20
        tax=(5/100)*base_bill
        print(f"------------BILL------------")
        print(f"{'Name':<20} {name}")
        print(f"{'Unit Consumed':<20} {units_consumed}")
        print(f"{'Base Bill':<20} {base_bill}")
        print(f"------------------------------")
        print(f"{'Tax(%5):':<20} {tax}")
        print(f"{'Total Bill':<20} {base_bill+tax}")
        print(f"------------------------------")
    elif (units_consumed>100) and (units_consumed<=200):
        first_hundred=100*20
        second_hundred=(units_consumed-100)*30
        base_bill=first_hundred+second_hundred
        tax=(5/100)*base_bill
        print(f"------------BILL------------")
        print(f"{'Name':<20} {name}")
        print(f"{'Unit Consumed':<20} {units_consumed}")
        print(f"{'First 100 Units':<20} {first_hundred}")
        print(f"{'Second 100 Units':<20} {second_hundred}")
        print(f"{'Base Bill':<20} {base_bill}")
        print(f"------------------------------")
        print(f"{'Tax(%5):':<20} {tax}")
        print(f"{'Total Bill':<20} {base_bill+tax}")
        print(f"------------------------------")
    elif units_consumed>200:
        first_hundred=100*20
        second_hundred=100*30
        remaining_units=abs(units_consumed-200)*40
        base_bill=first_hundred+second_hundred+remaining_units
        tax=(5/100)*base_bill
        print(f"------------BILL------------")
        print(f"{'Name':<20} {name}")
        print(f"{'Unit Consumed':<20} {units_consumed}")
        print(f"{'First 100 Units':<20} {first_hundred}")
        print(f"{'Second 100 Units':<20} {second_hundred}")
        print(f"{'Remaining Units':<20} {remaining_units}")
        print(f"{'Base Bill':<20} {base_bill}")
        print(f"------------------------------")
        print(f"{'Tax(%5):':<20} {tax}")
        print(f"{'Total Bill':<20} {base_bill+tax}")
        print(f"------------------------------")
else:
    print("Invalid Input:")
````

## 📝 Challenge 6: Grocery Billing System ⭐⭐⭐

**Task:** Create a point-of-sale (POS) billing system that calculates individual item totals, applies conditional discounts, computes government tax, and prints a polished invoice.

---

### 📋 Requirements

1. **User Inputs:** Prompt the user for both the **Price** and **Quantity** of three grocery items:
   * **Rice**
   * **Milk**
   * **Eggs**
2. **Calculations:**
   * **Item Totals:** Calculate the cost for each individual item ($\text{Price} \times \text{Quantity}$).
   * **Subtotal:** Sum the totals of all three items.
   * **Conditional Discount:** If the **Subtotal is greater than 5000**, apply a flat **10% discount** on the subtotal. Otherwise, the discount is 0.
   * **GST (Sales Tax):** Apply **18% GST** on the discounted subtotal (Subtotal minus Discount).
   * **Final Bill:** Calculate the final amount ($\text{Discounted Subtotal} + \text{GST}$).
3. **Output Formatting:** Print a neat, structured invoice displaying each item's total, the subtotal, the discount applied, the tax, and the final bill rounded to **2 decimal places**.

---

### 🧮 Mathematical Formulas

$$\text{Item Total} = \text{Price} \times \text{Quantity}$$

$$\text{Subtotal} = \text{Rice Total} + \text{Milk Total} + \text{Eggs Total}$$

$$\text{Discounted Subtotal} = \text{Subtotal} - \text{Discount}$$

$$\text{Final Bill} = \text{Discounted Subtotal} \times 1.18$$

---

### 📥 Example Output Formatting

```text
=======================================
           GROCERY INVOICE             
=======================================
Rice Total:        Rs. 3,200.00
Milk Total:        Rs. 1,500.00
Eggs Total:        Rs.    600.00
---------------------------------------
Subtotal:          Rs. 5,300.00
10% Discount:     -Rs.   530.00
18% GST:          +Rs.   858.60
---------------------------------------
TOTAL BILL:        Rs. 5,628.60
=======================================

````python
#Input
rice_quantity=float(input("Enter Quantity of Rice Bought:"))
rice_price=float(input("Enter Price of Rice:"))
milk_quantity=float(input("Enter Quantity of Milk Bought(Liters):"))
milk_price=float(input("Enter Price of Milk/Liter:"))
egg_quantity=int(input("Enter Quantity of Eggs Bought:"))
egg_price=float(input("Enter Price of Egg:"))
if ((rice_quantity>=0 and rice_price>=0) and (milk_quantity>=0 and milk_price>=0) and (egg_quantity>=0 and egg_price>=0)):
    rice_total=rice_quantity*rice_price
    milk_total=milk_quantity*milk_price
    eggs_total=egg_price*egg_quantity
    sub_total=rice_total+milk_total+eggs_total
    gst=(18/100)*sub_total
    if sub_total>=5000:
        discount=(10/100)*sub_total
        discount_subtotal=sub_total-discount
        total_price=discount_subtotal+gst
        print(f"==============================")
        print(f"        GROCERY INVOICE       ")
        print(f"==============================")
        print(f"{'Rice Total':<20} {rice_total:.2f}")
        print(f"{'Milk Total':<20} {milk_total:.2f}")
        print(f"{'Eggs Total':<20} {eggs_total:.2f}")
        print(f"------------------------------")
        print(f"{'Sub Total':<20} {sub_total:.2f}")
        print(f"{'10% Discount':<20} -{discount:.2f}")
        print(f"{'18 GST':<20} +{gst:.2f}")
        print(f"------------------------------")
        print(f"{'TOTAL BILL':<20} {total_price:.2f}")
        print(f"==============================")
    else:
        discount=0
        discount_subtotal=sub_total-discount
        total_price=discount_subtotal+gst
        print(f"==============================")
        print(f"        GROCERY INVOICE       ")
        print(f"==============================")
        print(f"{'Rice Total':<20} {rice_total:.2f}")
        print(f"{'Milk Total':<20} {milk_total:.2f}")
        print(f"{'Eggs Total':<20} {eggs_total:.2f}")
        print(f"------------------------------")
        print(f"{'Sub Total':<20} {sub_total:.2f}")
        print(f"{'10% Discount':<20} -{discount:.2f}")
        print(f"{'18 GST':<20} +{gst:.2f}")
        print(f"------------------------------")
        print(f"{'TOTAL BILL':<20} {total_price:.2f}")
        print(f"==============================")
else:
    print("Invalid Input.")
````

## 📝 Challenge 7: Age Calculator ⭐⭐⭐

### 📋 Requirements

   * **Age in Months:** Convert the age in years to months (using **12 months per year**).
   * **Approximate Age in Days:** Convert the age in years to days (using **365 days per year**).
3. **Output Formatting:** Print a clean summary displaying the calculated age in all three units.

---

### 🧮 Formulas

$$\text{Age in Years} = \text{Current Year} - \text{Birth Year}$$

## 📝 Challenge 8: Salary Calculator ⭐⭐⭐

**Task:** Create a payroll program that takes a basic salary, calculates allowances and taxes, and outputs a complete breakdown of gross and net pay.

---

### 📋 Definitions & Requirements

1. **User Input:** Prompt the user to enter their **Basic Salary** (float).
2. **Calculations:**
   * **House Rent Allowance (HRA):** 20% of the Basic Salary.
   * **Medical Allowance:** 10% of the Basic Salary.
   * **Gross Salary:** The sum of the Basic Salary and both allowances.
   * **Tax Deduction:** 8% of the calculated **Gross Salary**.
   * **Net Salary:** The final take-home pay after deducting tax from the Gross Salary.
3. **Output Formatting:** Print a clean pay stub showing each component. Round all monetary values to exactly **2 decimal places**.

---

### 🧮 Mathematical Structure

$$\text{Gross Salary} = \text{Basic Salary} + (0.20 \times \text{Basic}) + (0.10 \times \text{Basic}) = \text{Basic Salary} \times 1.30$$

$$\text{Tax Deduction} = \text{Gross Salary} \times 0.08$$

$$\text{Net Salary} = \text{Gross Salary} - \text{Tax Deduction}$$

---

### 📥 Example Walkthrough

#### Expected Input:
* **Basic Salary:** `100,000`

#### Expected Calculations & Output:
* **House Rent (20%):** `20,000.00`
* **Medical (10%):** `10,000.00`
* **Gross Salary:** `130,000.00`
* **Tax (8% of Gross):** `10,400.00`
* **Net Salary:** `119,600.00`

```text
========================================
               PAY SLIP                 
========================================
Basic Salary:       Rs. 100,000.00
House Allowance:    Rs.  20,000.00
Medical Allowance:  Rs.  10,000.00
----------------------------------------
GROSS SALARY:       Rs. 130,000.00
Tax Deduction (8%): Rs.  10,400.00
----------------------------------------
NET TAKE-HOME PAY:  Rs. 119,600.00
========================================

````python
#Input
basic_salary=float(input("Enter Your Basic Salary:"))
#Calculation
if basic_salary>0:
    house_allowance=basic_salary*(20/100)
    medical_allowance=basic_salary*(10/100)
    gross_salary=basic_salary+house_allowance+medical_allowance
    tax=gross_salary*(8/100)
    net_salary=gross_salary-tax
    #Output
    print(f"=====================================")
    print(f"              PAY SLIP               ")
    print(f"=====================================")
    print(f"{'Basic Salary:':<20} Rs.{basic_salary:.2f}")
    print(f"{'House Allowance:':<20} Rs.{house_allowance:.2f}")
    print(f"{'Medical Allowance:':<20} Rs.{medical_allowance:.2f}")
    print(f"-------------------------------------")
    print(f"{'GROSS SALARY:':<20} Rs.{gross_salary:.2f}")
    print(f"{'Tax Deduction (%8):':<20} Rs.{tax:.2f}")
    print(f"-------------------------------------")
    print(f"{'NET SALARY:':<20} Rs.{net_salary:.2f}")
    print(f"=====================================")
else:
    print("Invalid  Salary Input.")
````

## 📝 Challenge 9: Unit Converter ⭐⭐⭐

**Task:** Create an interactive, menu-driven program that allows users to select a conversion type from a list of options, input their value, and view the converted result.

---

### 📋 Requirements

1. **Interactive Menu:** Display a numbered list of conversion options to the user:
   1. Kilometer $\rightarrow$ Miles
   2. Miles $\rightarrow$ Kilometer
   3. Kilograms $\rightarrow$ Pounds
   4. Pounds $\rightarrow$ Kilograms
   5. Celsius $\rightarrow$ Fahrenheit
   6. Fahrenheit $\rightarrow$ Celsius
2. **User Selection:** Prompt the user to enter their choice (e.g., numbers `1` through `6`).
3. **Value Input:** Prompt the user to enter the numeric value they wish to convert.
4. **Conversion Execution:** Calculate the result based on their menu choice using the standard conversion rates below.
5. **Output & Error Handling:** Print the converted value rounded to **2 decimal places**. Display an error message if the user enters an invalid menu option.

---

### 🧮 Conversion Reference Table

| Option | Direction | Formula / Rate |
| :--- | :--- | :--- |
| **1** | Kilometer $\rightarrow$ Miles | $\text{Miles} = \text{Kilometers} \times 0.621371$ |
| **2** | Miles $\rightarrow$ Kilometer | $\text{Kilometers} = \text{Miles} \times 1.60934$ |
| **3** | Kilograms $\rightarrow$ Pounds | $\text{Pounds} = \text{Kilograms} \times 2.20462$ |
| **4** | Pounds $\rightarrow$ Kilograms | $\text{Kilograms} = \text{Pounds} \times 0.453592$ |
| **5** | Celsius $\rightarrow$ Fahrenheit | $F = (C \times \frac{9}{5}) + 32$ |
| **6** | Fahrenheit $\rightarrow$ Celsius | $C = (F - 32) \times \frac{5}{9}$ |

---

### 📥 Example Interface

```text
=======================================
            UNIT CONVERTER             
=======================================
1. Kilometer -> Miles
2. Miles -> Kilometer
3. Kilograms -> Pounds
4. Pounds -> Kilograms
5. Celsius -> Fahrenheit
6. Fahrenheit -> Celsius
---------------------------------------
Enter choice (1-6): 3
Enter weight in Kilograms: 75

Result: 75.00 kg is equal to 165.35 lbs

````python
while True:
    print(f"==========================================")
    print(f"            UNIT CONVERTER                ")
    print(f"==========================================")
    print("1. Kilometer-> Miles\n2. Miles-> Kilometer\n3. Kilograms-> Pounds\n4. Pounds-> Kilograms\n5. Celsius-> Fahrenheit\n6. Fahrenheit-> Celsius\n")
    choice=int(input("Enter Choice:"))
    if choice==1:
        distance=float(input("Enter Distance in Kilometer:"))
        conversion=distance*0.621371
        print(f"---------------------------------------")
        print(f"Result: {distance:.2f} km is equal to {conversion:.2f} miles")
        print(f"=======================================")
    elif choice==2:
        distance=float(input("Enter Distance in Miles:"))
        conversion=distance*1.60934
        print(f"---------------------------------------")
        print(f"Result: {distance:.2f} miles is equal to {conversion:.2f} km")
        print(f"=======================================")
    elif choice==3:
        weight=float(input("Enter Weight in Kilograms:"))
        conversion=weight*2.20462
        print(f"---------------------------------------")
        print(f"Result: {weight:.2f} kg is equal to {conversion:.2f} lbs")
        print(f"=======================================")
    elif choice==4:
        weight=float(input("Enter Weight in pounds:"))
        conversion=weight*0.453592
        print(f"---------------------------------------")
        print(f"Result: {weight:.2f} lbs is equal to {conversion:.2f} kg")
        print(f"=======================================")
    elif choice==5:
        temperature=float(input("Enter temperature in Celsius:"))
        conversion=(temperature*9/5)+32
        print(f"---------------------------------------")
        print(f"Result: {temperature:.2f} Celsius is equal to {conversion:.2f} Fahrenheit")
        print(f"=======================================")
    elif choice==6:
        temperature=float(input("Enter temperature in Fahrenheit:"))
        conversion=(temperature-32)*5/9
        print(f"---------------------------------------")
        print(f"Result: {temperature:.2f} Fahrenheit is equal to {conversion:.2f} Celsius")
        print(f"=======================================")
    else:
        print("Invalid Input.Choose a number between 1-6")
    ans=(str(input("Do you want to continue Y/n?"))).lower()
    if ans!="y":
        break


````
