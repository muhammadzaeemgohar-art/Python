### 📝 Challenge 1: Temperature Converter 

**Task:** Write a program that converts temperatures seamlessly between **Celsius**, **Fahrenheit**, and **Kelvin**.

---

### 📋 Requirements

1. **Temperature Input:** Prompt the user to enter a numeric temperature value.
2. **Current Unit:** Prompt the user to enter the source unit (`C`, `F`, or `K`).
3. **Target Unit:** Prompt the user to enter the desired unit (`C`, `F`, or `K`).
4. **Conversion Logic:** Correctly handle all 6 possible conversion combinations.
5. **Precision:** Print the final answer rounded to exactly **2 decimal places**.
6. **Error Handling:** Gracefully handle and report any invalid unit inputs.

---

### 💡 Quick Reference Formulas

| Conversion | Formula |
| :--- | :--- |
| **Celsius to Fahrenheit** | $F = (C \times \frac{9}{5}) + 32$ |
| **Celsius to Kelvin** | $K = C + 273.15$ |
| **Fahrenheit to Celsius** | $C = (F - 32) \times \frac{5}{9}$ |

<div style="padding: 15px; border-left: 5px solid #2196F3; background-color: #eef7fe; color: #0d47a1; border-radius: 4px;">
    <strong>ℹ️ Tip:</strong> Don't forget to test edge cases, such as entering lowercase letters (<code>c</code>, <code>f</code>, <code>k</code>) or absolute zero!
</div>

---
### Solution

```python
current_temp=float(input("Enter Temperature:"))
current_unit=(str(input("Enter the source unit C,F or K:"))).capitalize()
target_unit=(str(input("Enter target unit C,F or K:"))).capitalize()
#Celcius to Farhenhit
if ((current_unit=="C") and (target_unit =="F")):
        converted_temp=((current_temp)*(9/5))+32
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Celcius to Kelvin
elif ((current_unit=="C") and (target_unit=="K")):
    converted_temp=current_temp+273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Farhenheit to Celcius
elif ((current_unit=="F") and (target_unit=="C")):
    if current_temp==32:
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is 0")
    else:
        converted_temp=(current_temp-32)*(5/9)
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Farhenheit to Kelvin
elif ((current_unit=="F") and (target_unit=="K")):
    converted_temp=(current_temp-32)*(5/9)+273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#kelvin to Celcius
elif ((current_unit=="K") and (target_unit=="C")):
    converted_temp=current_temp-273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Kelvin to Farhenheit
elif ((current_unit=="K") and (target_unit=="F")):
    converted_temp=(current_temp-273.15)*(9/5)+32
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#For Same Units
elif (current_unit==target_unit):
    print(f"The Temperature in {current_unit} is {current_temp}")
else:
    print("Invalid Input! Try Again.")
```

## 📝 Challenge 2: BMI Calculator ⭐

**Task:** Write a program that calculates a user's Body Mass Index (BMI) based on their height and weight, and determines their weight category.

---

### 📋 Requirements

1. **User Inputs:** Prompt the user for the following three inputs:
   * **Name** (Text)
   * **Weight** in kilograms (Numeric)
   * **Height** in meters (Numeric)
2. **BMI Calculation:** Calculate the BMI using the formula below.
3. **Output Formatting:** Print a clean summary displaying:
   * The user's **Name**
   * Their calculated **BMI** rounded to exactly **2 decimal places**
   * Their official **BMI Category** based on the classification table below.

---

### 🧮 Formula & Classifications

$$BMI = \frac{\text{weight}}{\text{height}^2}$$

| BMI Range | Category |
| :--- | :--- |
| **Less than 18.5** | Underweight |
| **18.5 – 24.9** | Normal |
| **25.0 – 29.9** | Overweight |
| **30.0 or more** | Obese |

<div style="padding: 15px; border-left: 5px solid #4CAF50; background-color: #f1f8e9; color: #33691e; border-radius: 4px;">
    <strong>ℹ️ Tip:</strong> Make sure your height input handles decimal values (like <code>1.75</code>) and weight handles floats (like <code>68.5</code>) to ensure calculation accuracy!
</div>

---
### 💻Solution

```python
#Inputs
name=str(input("Enter Your Name:"))
weight=float(input("Enter Your Weight in kilograms:"))
height=float(input("Enter Your Height in meters:"))
#Calculation
bmi=weight/(height)**2
#classification
if bmi<18.5:
    category="Underweight"
elif bmi>=18.5 and bmi<=24.9:
    category="Normal"
elif bmi>=25.0 and bmi<=29.9:
    category="Overweight"
else:
    category="obese"
#Results
print(f"{'Name':<12} {name}")
print(f"{'BMI':<12} {bmi:.2f}")
print(f"{'Category':<12} {category}")
```
