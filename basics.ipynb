### 📝 Challenge 1: Temperature Converter 

**Task:** Write a program that converts temperatures seamlessly between **Celsius**, **Fahrenheit**, and **Kelvin**.

---

### 📋 Requirements

1. **Temperature Input:** Prompt the user to enter a numeric temperature value.
2. **Current Unit:** Prompt the user to enter the source unit (`C`, `F`, or `K`).
3. **Target Unit:** Prompt the user to enter the desired unit (`C`, `F`, or `K`).
4. **Conversion Logic:** Correctly handle all 6 possible conversion combinations.
5. **Precision:** Print the final answer rounded to exactly **2 decimal places**.
6. **Error Handling:** Gracefully handle and report any invalid unit inputs.

---

### 💡 Quick Reference Formulas

| Conversion | Formula |
| :--- | :--- |
| **Celsius to Fahrenheit** | $F = (C \times \frac{9}{5}) + 32$ |
| **Celsius to Kelvin** | $K = C + 273.15$ |
| **Fahrenheit to Celsius** | $C = (F - 32) \times \frac{5}{9}$ |

### 💻 My Code Solution

```python
current_temp=float(input("Enter Temperature:"))
current_unit=(str(input("Enter the source unit C,F or K:"))).capitalize()
target_unit=(str(input("Enter target unit C,F or K:"))).capitalize()
#Celcius to Farhenhit
if ((current_unit=="C") and (target_unit =="F")):
        converted_temp=((current_temp)*(9/5))+32
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Celcius to Kelvin
elif ((current_unit=="C") and (target_unit=="K")):
    converted_temp=current_temp+273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Farhenheit to Celcius
elif ((current_unit=="F") and (target_unit=="C")):
    if current_temp==32:
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is 0")
    else:
        converted_temp=(current_temp-32)*(5/9)
        print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Farhenheit to Kelvin
elif ((current_unit=="F") and (target_unit=="K")):
    converted_temp=(current_temp-32)*(5/9)+273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#kelvin to Celcius
elif ((current_unit=="K") and (target_unit=="C")):
    converted_temp=current_temp-273.15
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#Kelvin to Farhenheit
elif ((current_unit=="K") and (target_unit=="F")):
    converted_temp=(current_temp-273.15)*(9/5)+32
    print(f"The Temperature in {current_unit} is {current_temp} and in {target_unit} is {converted_temp}")
#For Same Units
elif (current_unit==target_unit):
    print(f"The Temperature in {current_unit} is {current_temp}")
else:
    print("Invalid Input! Try Again.")
```
