## 📝 Challenge 01: Smart Grade Calculator ⭐⭐⭐

**Task:** Write a grade classification program that inputs a student's score, validates that the score is within a realistic range, assigns a letter grade, and determines their overall pass/fail status.

---

### 📋 Requirements

1. **User Input:** Prompt the user to enter a numeric score (integer or float) between `0` and `100`.
2. **Data Validation:** Before performing any grade evaluation, check if the input is valid. If the score is **below 0** or **above 100**, reject it with an error message and terminate or skip calculation.
3. **Grading Logic:** Assign a letter grade based on the following scale:
   * **90 – 100:** A
   * **80 – 89:** B
   * **70 – 79:** C
   * **60 – 69:** D
   * **Below 60:** F
4. **Status Check:** Determine the completion status. Any grade of **D or above is a Pass**, while a grade of **F is a Fail**.
5. **Output Formatting:** Print a clean report showing the student's score, their letter grade, and whether they passed or failed.

---

### 📥 Example Walkthrough

#### Case A: Valid Entry (Pass)
* **Input Score:** `78.5`
* **Expected Output:**
  ```text
  Score: 78.50%
  Grade: C
  Status: PASSED
  Error: Invalid score entered. Marks must be between 0 and 100.

````python
marks=float(input("Enter your marks(0-100):"))
if marks>=0 and marks<=100:
    if marks>=90 and marks<=100:
        grade="A"
    elif marks>=80 and marks<=89:
        grade="B"
    elif marks>=70 and marks<=79:
        grade="C"
    elif marks>=60 and marks<=69:
        grade="D"
    elif marks<60:
        grade="F"
    if grade=="F":
        status="FAIL"
    else:
        status="PASSED"
    print(f"Score: {marks}")
    print(f"Grade: {grade}")
    print(f"Status: {status}")
else:
     print("Invalid score entered. Marks must be between 0 and 100.")

 
````

## 📝 Challenge 02: ATM PIN Verification ⭐⭐⭐

**Task:** Create a basic ATM system simulator that verifies a user's security PIN and processes cash withdrawals based on account balances and safety parameters.

---

### 📋 Requirements

1. **System Constants:** Pre-program a fixed, valid security PIN (e.g., `1234`) and establish a starting account balance of **Rs. 50,000**.
2. **User Input:** Prompt the user to enter two inputs:
   * **Security PIN** (Integer)
   * **Withdrawal Amount** (Float/Integer)
3. **Verification & Transaction Engine:**
   * **PIN Validation:** Check the input PIN first. If it does not match your system constant, immediately print `"Access Denied"` and cancel the transaction.
   * **Amount Validation:** If the PIN is correct, check the withdrawal amount. Reject any negative amounts (e.g., less than or equal to `0`).
   * **Balance Validation:** Ensure the requested cash amount does not exceed the available account balance. 
     * If `withdrawal > balance`, print `"Insufficient Funds"`.
     * If `withdrawal <= balance`, deduct the funds and approve the payout.
4. **Output Formatting:** For successful transactions, print a clean summary revealing the withdrawn sum and the remaining account balance.

---

### 📥 Example Walkthroughs

#### Case A: Authentication Failure
* **Input PIN:** `9999`
* **Expected Output:**
```markdown
Access Denied
========================================
          TRANSACTION SUCCESS            
========================================
Withdrawn Amount:   Rs. 15,500.00
Remaining Balance:  Rs. 34,500.00
========================================

````python
sec_pin=1234
account_balance=50000
pin=int(input("Enter Security Pin:"))
if pin>=0:
    if pin==sec_pin:
        print(F"Login Successful")
        withdrawn_amount=int(input("Enter Amount of Money you want to Withdraw:"))
        if withdrawn_amount>0: 
            if account_balance>=withdrawn_amount:
                remaining_balance=account_balance-withdrawn_amount
                print(f"================================================")
                print(f"              TRANSACTION SUCCESS               ")
                print(f"================================================")
                print(f"Withdrawn Amount:     Rs. {withdrawn_amount}")
                print(f"Remaining Balance:    Rs. {remaining_balance}")
                print(f"================================================")
            else:
                print("Insufficent Balance.")
        else:
            print("Withdrawal amount must be greater than 0.Try Again.")
    else:
        print(f"Access Denied.Incorrect Password")
else:
    print("Invalid Format.Pin cannot be negative.")
````

## 📝 Challenge 03: Age Eligibility Checker ⭐⭐⭐

**Task:** Write a verification program that evaluates a user's age and nationality to determine their legal eligibility for critical identity documents and civic rights.

---

### 📋 Requirements

1. **User Inputs:** Prompt the user to provide two parameters:
   * **Age** (Integer)
   * **Nationality** (String, e.g., `"Pakistani"` or `"Other"`)
2. **Data Validation:** Validate the age input immediately. Reject any entry that is **less than 0** or physically unrealistic (e.g., above 120) with an appropriate error message.
3. **Eligibility Rules:** Evaluate the following documentation criteria:
   * **CNIC:** Eligible if age is **18 or above**.
   * **Voting Rights:** Eligible if age is **18 or above** AND nationality is **"Pakistani"**.
   * **Driving License:** Eligible if age is **18 or above**.
   * **Passport:** Eligible at **any age** (no minimum requirement).
4. **Output Formatting:** Print a clean checklist showing which items the user is currently eligible to apply for.

---

### 📥 Example Walkthroughs

#### Case A: Pakistani Adult (Full Eligibility)
* **Input Age:** `21`
* **Input Nationality:** `Pakistani`
* **Expected Output:**
```markdown
========================================
       ELIGIBILITY REPORT CARD          
========================================
✅ Eligible for Passport
✅ Eligible for CNIC
✅ Eligible for Driving License
✅ Eligible to Vote in Pakistan
========================================

````python
age=int(input("Enter your age:"))
nationality=str(input("Enter your Nationality (Pakistani or Other):")).lower()
if (nationality=="pakistani" or nationality=="other") and age>=0 and age<=120:
    if age>=18 and nationality=="pakistani":
        cnic="Eligible"
        voting="Eligible"
        driving="Eligible"
    elif age>=18 and nationality=="other":
        cnic="Eligible"
        voting="Not Eligible"
        driving="Eligible"
    else:
        cnic="Not Eligible"
        voting="Not Eligible"
        driving="Not Eligible"
    print(f"==================================")
    print(f"       ELIGIBLITY REPORT CARD     ")
    print(f"==================================")
    print(f"{'Eligible'} for Passport.")
    print(f"{cnic} for CNIC.")
    print(f"{driving} for Driving License")
    print(f"{voting} for Voting in Pakistan")
    print(f"==================================")
else:
    print("Invalid Input.Incorrect Age or Nationality.")
````

## 📝 Challenge 04: Income Tax Calculator ⭐⭐⭐

**Task:** Create a tax assessment tool that calculates a user's progressive annual income tax bracket, computes the specific tax liabilities, and tracks the net take-home salary.

---

### 📋 Requirements

1. **User Input:** Prompt the user to enter their **Annual Salary** (Float/Integer).
2. **Data Validation:** Validate the input immediately. Reject any entry that is **less than 0** with an appropriate error message.
3. **Tax Bracket Margins:** Calculate the tax liability according to the following progressive rules:
   * **Up to 500,000:** 0% tax
   * **500,001 to 1,000,000:** 5% tax
   * **1,000,001 to 2,000,000:** 10% tax
   * **Above 2,000,000:** 15% tax
4. **Output Formatting:** Print a clean summary displaying the original salary, the corresponding tax percentage rate, the calculated tax amount, and the net salary remaining after tax. Round all numeric calculations to exactly **2 decimal places**.

---

### 📥 Example Walkthroughs

#### Case A: Low-Income Bracket (Tax-Exempt)
* **Input Salary:** `450000`
* **Expected Output:**
```markdown
========================================
             TAX STATEMENT              
========================================
Gross Salary:     Rs. 450,000.00
Tax Rate Applied: 0%
Tax Amount:       Rs. 0.00
----------------------------------------
Net Salary:       Rs. 450,000.00
========================================

````python
annual_salary=float(input("Enter Your Annual Salary:"))
if annual_salary>=0:
    if annual_salary<=500000:
        tax_rate=0
        tax_amount=0
        net_salary=annual_salary-tax_amount
    elif annual_salary>500000 and annual_salary<=1000000:
        tax_rate=5
        tax_amount=annual_salary*(5/100)
        net_salary=annual_salary-tax_amount
    elif annual_salary>1000000 and annual_salary<=2000000:
        tax_rate=10
        tax_amount=annual_salary*(10/100)
        net_salary=annual_salary-tax_amount
    elif annual_salary>2000000:
        tax_rate=15
        tax_amount=annual_salary*(15/100)
        net_salary=annual_salary-tax_amount
    print(f"==================================")
    print(f"          TAX STATEMENT           ")
    print(f"==================================")
    print(f"Gross Salary:     Rs. {annual_salary}")
    print(f"Tax Rate Applied: {tax_rate}%")
    print(f"Tax Amount:     Rs. {tax_amount}")
    print(f"----------------------------------")
    print(f"Net Salary:     Rs. {net_salary}")
    print(f"==================================")
else:
    print("Invalid Input.Reenter Salary")
````
